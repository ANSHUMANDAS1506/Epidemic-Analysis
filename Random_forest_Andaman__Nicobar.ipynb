```python
ts = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Project 1/Andaman.csv')
```

```python
ts_Andaman = ts[ts['State'] ==  'Andaman and Nicobar Islands']
```

```python
ts_Andaman = ts_Andaman.drop(columns=['Negative'])
```

```python
ts_Andaman_clean = ts_Andaman.dropna()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from matplotlib import pyplot as plt

# Load and prepare the data
data = ts_Andaman_clean[['Date', 'Positive']].copy()  # Adjust column names if needed
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Positive'].values.reshape(-1, 1))

# Prepare the dataset for training and testing
def prepare_data(series, n_steps):
    X, y = [], []
    for i in range(len(series) - n_steps):
        X.append(series[i:i + n_steps].flatten())
        y.append(series[i + n_steps])
    return np.array(X), np.array(y)

n_steps = 30
X, y = prepare_data(scaled_data, n_steps)

# Split the data into training and testing sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Train the Random Forest model
model = RandomForestRegressor(n_estimators=200, random_state=42, max_depth=15)  # Adjusted parameters for potential improvement
model.fit(X_train, y_train)

# Generate forecasts for the next 30 days
forecast_input = scaled_data[-n_steps:].flatten()
forecast = []

for _ in range(30):
    prediction = model.predict(forecast_input.reshape(1, -1))
    forecast.append(prediction[0])
    forecast_input = np.append(forecast_input[1:], prediction)

# Inverse scale the forecast
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# Plot the actual data and forecast
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Positive'], label="Actual", color='blue')
future_dates = pd.date_range(data.index[-1], periods=30 + 1, freq='D')[1:]
plt.plot(future_dates, forecast, label="Forecast", color='red', linestyle='--')

# Add labels, title, and legend
plt.title('Actual vs Forecasted Positive Cases')
plt.xlabel("Date")
plt.ylabel("Positive Cases")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

# Predict on the test set
y_pred_scaled = model.predict(X_test)

# Inverse scale the predictions and actual values
y_pred = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate error metrics
mse = mean_squared_error(y_actual, y_pred)
mae = mean_absolute_error(y_actual, y_pred)
mape = mean_absolute_percentage_error(y_actual, y_pred)

# Print the results
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Mean Absolute Percentage Error (MAPE):", mape * 100, "%")
```

```python
# Calculate MAPE
mape = mean_absolute_percentage_error(y_test, y_pred) * 100  # MAPE in percentage

# Calculate accuracy as (100% - MAPE)
accuracy = 100 - mape

# Print accuracy
print(f'Model Accuracy: {accuracy:.2f}%')
```

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted values
plt.figure(figsize=(14, 7))
plt.plot(data.index[-len(y_actual):], y_actual, label='Actual', color='blue', linewidth=2)
plt.plot(data.index[-len(y_pred):], y_pred, label='Predicted', color='red', linestyle='--')

# Set labels and title
plt.title('Actual vs Predicted Positive Cases')
plt.xlabel('Date')
plt.ylabel('Positive Cases')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
